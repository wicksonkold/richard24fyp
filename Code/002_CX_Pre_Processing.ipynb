```python
def read_first_csv_in_folder(folder_path):
    # List all files in the directory and filter out all CSV files
    files = [file for file in os.listdir(folder_path) if file.endswith('.csv')]
    # Sort the files to read the first one (optional, depending on your need)
    files.sort()
    for file in files:
        file_path = os.path.join(folder_path, file)
        # Check if the file is not empty
        if os.path.getsize(file_path) > 0:
            df = pd.read_csv(file_path)
            return df
    return None  # Return None if no non-empty CSV files are found

# List of folders
subfolders = [
    r'C:\Users\User\Desktop\Final Year Project\Code\Processed_CX\01_Columns_Type_Classification\Result_2\Style_1',
    r'C:\Users\User\Desktop\Final Year Project\Code\Processed_CX\01_Columns_Type_Classification\Result_2\Style_2',
    r'C:\Users\User\Desktop\Final Year Project\Code\Processed_CX\01_Columns_Type_Classification\Result_2\Style_3',
    r'C:\Users\User\Desktop\Final Year Project\Code\Processed_CX\01_Columns_Type_Classification\Result_2\Style_4',
    r'C:\Users\User\Desktop\Final Year Project\Code\Processed_CX\01_Columns_Type_Classification\Result_2\Style_5',
    r'C:\Users\User\Desktop\Final Year Project\Code\Processed_CX\01_Columns_Type_Classification\Result_2\Style_6'
]
```

### Style 2

```python
df = None
for i in range(3):
    # Input and output directories
    num = i + 1
    input_folder = fr'C:\Users\User\Desktop\Final Year Project\Code\Processed_CX\01_Columns_Type_Classification\Result_{num}\Style_2'
    output_folder = fr'C:\Users\User\Desktop\Final Year Project\Code\Processed_CX\02_Transform_Data_Frame\Result_{num}'
    
    # Ensure output directory exists
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Use glob to list all CSV files in the directory
    csv_files = glob.glob(os.path.join(input_folder, '*.csv'))


    # Process each CSV file in the directory
    for file_path in csv_files:
        print(file_path)
        # Read the CSV file into a DataFrame
        df = pd.read_csv(file_path)
        df.dropna(subset=['Offset'], inplace=True)
        
        # Get the base name of the file (without the path)
        file_name = os.path.basename(file_path)

        # Add a new column with the filename
        df['Filename'] = file_name

        def create_adjusted_landing_time(df):
            import datetime
            # Ensure 'Date-Time at Touchdown' column is in datetime format
            df['Date-Time at Touchdown'] = pd.to_datetime(df['Date-Time at Touchdown'], errors='coerce')

            # Check if there are any non-null values in the 'Date-Time at Touchdown' column
            if df['Date-Time at Touchdown'].notnull().sum() == 0:
                # If all values are null, set 'adjusted landing time' to null for all rows
                df['adjusted landing time'] = pd.NaT
            else:
                # Find the last non-null time in the 'Date-Time at Touchdown' column
                last_valid_time = df['Date-Time at Touchdown'].dropna().iloc[-1]

                # Calculate the number of valid time entries
                valid_times_count = df['Date-Time at Touchdown'].notnull().sum()

                # Create a list of decremented times starting from the last valid time
                adjusted_times = [last_valid_time - pd.Timedelta(seconds=x) for x in range(valid_times_count)]

                # Initialize the 'adjusted landing time' column with null values
                df['adjusted landing time'] = pd.NaT

                # Assign the decremented times to the corresponding last rows in 'adjusted landing time' column
                df['adjusted landing time'][-valid_times_count:] = adjusted_times[::-1]

            return df

        # Apply the function to each row
        df = create_adjusted_landing_time(df)

        # Define output file path
        output_file_path = os.path.join(output_folder, file_name.replace('.csv', '.xlsx'))

        # Save the updated DataFrame to an Excel file
        df.to_excel(output_file_path, index=False)
```
